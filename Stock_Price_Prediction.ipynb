```python
cols_plot = ['Open', 'Close', 'High','Low']
axes = df[cols_plot].plot(figsize=(12, 10), subplots=True)
for ax in axes:
    ax.set_ylabel('Daily trade')
```

```python
df.hist(figsize=(13, 13),column=['%Deliverble'])
```

```python
variation = df.groupby('Year')[['Turnover']].plot(figsize=(13,13))
```

```python
max_high=df.groupby('Year').sum()[['High']].sort_values(by='High',ascending=False)
max_high
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,5))
sns.heatmap(df[['High','Low','Volume','Turnover','Deliverable Volume','%Deliverble']].corr(),annot=True)
plt.show()
```

```python
one = [df['High'].max(),df['Low'].min(),df['Turnover'].max()]
print(one)
two = [df2['High'].max(),df2['Low'].min(),df2['Turnover'].max()]
print(two)
three = [df3['High'].max(),df3['Low'].min(),df3['Turnover'].max()]
print(three)
four = [df4['High'].max(),df4['Low'].min(),df4["Turnover"].max()]
print(four)


highest_to=max(one[2],two[2],three[2],four[2])
print("Max turnover was achieved by - ")
print(highest_to)
```
